```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns

#for storing the plots in the notebook document.
%matplotlib inline

# sns.set_style("dark")
plt.rcParams['figure.figsize'] = (16, 8)
```

```python
matches = pd.read_csv('/content/IPL_Matches_2008_2022.csv')
```

```python
print(matches.shape)
```

<h2>Total number of matches played</h2>

```python
print("Total matches played till date:",matches["ID"].count())
```

<h2>Total number of IPL Seasons</h2>

```python
print("Total seasons played till date:",len(matches['Season'].unique()))
```

<h2>Total number of matches played in each season</h2>

```python
c = sns.countplot(x='Season', data=matches)
c.set_xlabel('Season')
c.set_ylabel('Match Count')
plt.title("Number of matches played in each season",fontsize=20)
plt.show()
```

<h2>Maximum and minimum matches in a season</h2>

```python
matches_in_each_season = matches.groupby(['Season'])['ID'].count()
print('Most number of matches "{}" were played in "{}"\n'.format(matches_in_each_season.max(),matches_in_each_season.idxmax()))
print('Least number of matches "{}" were played in "{}"'.format(matches_in_each_season.min(),matches_in_each_season.idxmin()))
```

<h2>Total number of matches won by each team</h2>

```python
data = matches.WinningTeam.value_counts()
sns.barplot(y = data.index, x = data, orient='h')
plt.title("Number of matches won by each team",fontsize=20)
plt.show()
```

**ðŸ“ŒInference**: Mumbai Indians have the maximum matches followed by Chennai Super Kings.

```python
def add_labels(x,y):
  y = list(y.sort_values(ascending=False))
  for i in range(len(x)):
    plt.text(i,y[i],y[i],ha='center')
  return
```

<h2>Number of finals won by teams</h2>

```python
final_winners = matches[matches['MatchNumber'] == 'Final']
grouped_winners = final_winners.groupby(['WinningTeam'])['ID'].count().reset_index(name='No_of_finals_won')
grouped_winners_in_dec_order = grouped_winners.sort_values(by='No_of_finals_won',ascending=False)

colors=['green', 'orange', 'cyan', 'blue','red','pink','brown']

plt.bar(grouped_winners_in_dec_order['WinningTeam'],grouped_winners_in_dec_order['No_of_finals_won'],color=colors)
add_labels(grouped_winners_in_dec_order['WinningTeam'],grouped_winners_in_dec_order['No_of_finals_won'])
plt.title("Number of finals won by teams",fontsize=20)
plt.ylabel("Number of finals won")
plt.show()
```

<h2>Total number of matches played in each standium</h2>

```python
# count_of_matches_in_each_venue = matches.groupby(['Venue'])['ID'].count().reset_index(name='no_of_total_matches_played')
c = sns.countplot(x='Venue', data=matches)
c.set_xlabel('Venue')
c.set_ylabel('Number of matches played')
plt.title("Number of matches played in each stadium",fontsize=20)
c.set_xticklabels(c.get_xticklabels(), rotation=60, ha="right")
plt.tight_layout()
plt.show()
```

<h2>Top 10 stadiums with maximum matches</h2>

```python
count_of_matches_in_each_venue = matches.groupby(['Venue'])['Venue'].count().reset_index(name="num_of_match")
num_of_match_sorted_in_dec_order = sorted(list(count_of_matches_in_each_venue['num_of_match']),reverse=True)
plt.bar(count_of_matches_in_each_venue['Venue'],num_of_match_sorted_in_dec_order)
add_labels(count_of_matches_in_each_venue['Venue'],count_of_matches_in_each_venue['num_of_match'])
plt.title("Total Matches conducted at each stadium",fontsize=20)
plt.ylabel("Number of Matches")
plt.xticks(rotation=90)
plt.show()
```

```python
print("Total stadiums: ",len(count_of_matches_in_each_venue['Venue'].unique()))
print("Total matches:",sum(num_of_match_sorted_in_dec_order))
print("Matches conducted in top 10 stadiums:",sum(num_of_match_sorted_in_dec_order[:11]), "which is {:.2f}% of total matches.".format((sum(num_of_match_sorted_in_dec_order[:11])/sum(num_of_match_sorted_in_dec_order))*100) )
```

ðŸ“Œ**Inference**: <b>Among 49 stadiums, almost 60% of the matches have been conducted in only 10 stadiums. These should be evenly distributed across India</b> but this might have other consequences.

Note: In the above calculations I have ignored the fact that few matches were conducted in UAE too and assumed all to be conducted in India only.

<h2>Decison made after winning the toss</h2>

```python
toss_decision_series = matches['TossDecision'].value_counts()
labels = (np.array(toss_decision_series.index))
values = (np.array((toss_decision_series / toss_decision_series.sum())*100))

palette_color = sns.color_palette('dark')

fig, ax = plt.subplots()
patches, texts, pcts = ax.pie(values,labels=labels,colors=palette_color,autopct='%.1f%%',textprops={'size': 'x-large'})
plt.setp(pcts, color='white', fontweight='bold')
plt.title('Decison made after winning the toss',fontsize=20)
plt.show()
```

**ðŸ“ŒInference:** After winning the toss, a team tends to field first and chase the target.

<h2>Has Toss-winning helped in Match-winning?</h2>

```python
number_of_times_toss_winner_is_match_winner = matches[matches['TossWinner'] == matches['WinningTeam']]['ID'].count()
number_of_times_toss_looser_is_match_winner = matches[matches['TossWinner'] != matches['WinningTeam']]['ID'].count()

labels = ['percentage_of_times_toss_winner_is_match_winner','percentage_of_times_toss_looser_is_match_winner']
fig, ax = plt.subplots()
patches, texts, pcts = ax.pie([number_of_times_toss_winner_is_match_winner,number_of_times_toss_looser_is_match_winner],labels=labels,autopct='%.1f%%',textprops={'size': 'x-large'})
plt.setp(pcts, color='white', fontweight='bold')
ax.set_title("Toss win/loose vs match win",fontsize=20)
plt.tight_layout()
plt.show()
```

ðŸ“Œ**Inference**: Though winning toss gives teams an advantage but **it doesn't significantly implies that winning the toss helps in winning the game.**

<h2>Toss decision vs win/loss</h2>

```python
toss_decision_bat = matches[matches['TossDecision'] == 'bat']
toss_decision_bat_and_win =  toss_decision_bat[toss_decision_bat['TossWinner'] == toss_decision_bat['WinningTeam']]['ID'].count()

toss_decision_bat = matches[matches['TossDecision'] == 'bat']
toss_decision_bat_and_loose =  toss_decision_bat[toss_decision_bat['TossWinner'] != toss_decision_bat['WinningTeam']]['ID'].count()

toss_decision_field = matches[matches['TossDecision'] == 'field']
toss_decision_field_and_win =  toss_decision_field[toss_decision_field['TossWinner'] == toss_decision_field['WinningTeam']]['ID'].count()

toss_decision_field = matches[matches['TossDecision'] == 'field']
toss_decision_field_and_loose =  toss_decision_field[toss_decision_field['TossWinner'] != toss_decision_field['WinningTeam']]['ID'].count()

data = [toss_decision_bat_and_win,toss_decision_bat_and_loose,toss_decision_field_and_win,toss_decision_field_and_loose]
labels = ['toss_decision_bat_and_win','toss_decision_bat_and_loose','toss_decision_field_and_win','toss_decision_field_and_loose']
colors = sns.color_palette('pastel')[0:5]

plt.pie(x=data,labels=labels,colors=colors,autopct='%.0f%%')
plt.title('Toss decision vs win/loose',fontsize=20)
plt.show()
```

**ðŸ“ŒInference:** The chance of winning or loosing when a team choose to bat after winning the toss is approximately equal but *when the team fields after winning the toss, then the chnaces of winning the match is higher*.

<h2>Top 15 player of the match winners</h2>

```python
pom_winners = matches.groupby(['Player_of_Match'])['Player_of_Match'].count().reset_index(name="num_of_pom")
top_15_pom_winners = pom_winners.sort_values(by='num_of_pom',ascending=False).reset_index().head(15)
plt.bar(top_15_pom_winners['Player_of_Match'],top_15_pom_winners['num_of_pom'])
add_labels(top_15_pom_winners['num_of_pom'],top_15_pom_winners['num_of_pom'])
plt.title('Top 10 player of the match winners',fontsize=20)
plt.ylabel("Number of player of the match award")
plt.xlabel("Name of player")
plt.show()

top_15_pom_winners.columns = ['index','Player Name','#_Player_of_the_match_won']
top_15_pom_winners.drop('index',axis=1).style.background_gradient(cmap='PuBu')
```
